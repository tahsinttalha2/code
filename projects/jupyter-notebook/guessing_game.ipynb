```python
from random import randint
from IPython.display import clear_output

guessed = False
number = randint(0, 100)
guesses = 0

while not guessed:
    ans = input("Think of the number I've guessed.")
    guesses += 1
    if int(ans) > number:
        print("Yours one is bigger")
    elif int(ans) < number:
        print("Yours one is smaller")
    else:
        print("Perfect!")
        if (guesses > 1):
            print("It took you {} guesses.".format(guesses))
        else:
            print("You've got it on your first try!")
        break
```
